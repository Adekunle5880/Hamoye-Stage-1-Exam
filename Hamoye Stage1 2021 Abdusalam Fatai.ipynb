```python
# Importing libraries
import pandas as pd
import numpy as np
```

```python
# reading the csv data
fuel_data = pd.read_csv('https://github.com/WalePhenomenon/climate_change/blob/master/fuel_ferc1.csv?raw=true', error_bad_lines=False)
```

```python
# Getting the basic statistics of the numerical column
fuel_data.describe(include='all')
```

```python
#getting the sum of the missing values
fuel_data.isnull().sum()
```

```python
#grouping energy data by fuel_unit and counting each element in the column
fuel_data.groupby("fuel_unit")["fuel_unit"].count()
```

```python
# filling all empty column in fuel column the mode of the entries which is "mcf"
fuel_data[['fuel_unit']] = fuel_data[['fuel_unit']].fillna(value='mcf')
```

```python
#checking out null values
fuel_data.isnull().sum()
```

##### There is no more empty values

```python
# grouping by report_year and count
fuel_data.groupby("report_year")["report_year"].count()
```

```python
# grouping by fuel_type_code_pud1
fuel_data.groupby("fuel_type_code_pudl").first()
```

```python
# creating a new dataframe by getting 0-190000 rows of the data
fuel_df1 = fuel_data.iloc[0:19000].reset_index(drop=True)
```

```python
# creating a new dataframe by getting from 190000 rows downward of the data
fuel_df2 = fuel_data.iloc[19000:].reset_index(drop=True)
```

```python
# checking the lenght of the generated dataframe if it's equal. Assert check for the condition and continue
assert len(fuel_data) == (len(fuel_df1) + len(fuel_df2))
```

```python
# lose rows that do not match in both dataframe. 
pd.merge(fuel_df1, fuel_df2, how="inner")
```

```python
# return all rows in both dataframe
pd.merge(fuel_df1, fuel_df2, how="outer")
```

```python
# importing our visualization tools
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# creating a new dataframe
fuel_unit = pd.DataFrame({'unit':['BBL', 'GAL', 'GRAMSU', 'KGU', 'MCF', 'MMBTU',
                          'MWDTH', 'MWHTH', 'TON'],
                          'count':[7998, 84, 464, 110, 11354, 180, 95, 100, 8958]})
```

```python
# checking fuel unit on bar plot
plt.figure(figsize=(7,4))
#plt.xticks(rotation=90)

sns.barplot(data=fuel_unit, x='unit', y='count')
plt.xlabel('Fuel Unit')
```

```python
# Increasing the scale of the graph to see the values clearly
g = sns.barplot(data=fuel_unit, x='unit', y='count')
g.set_yscale("log")
g.set_ylim(1, 12000)
plt.xlabel('Fuel Unit')
```
